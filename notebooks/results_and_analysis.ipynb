Our logistic regression model performed best during the cross validation phase. That logistic regression would outperform our gradient boosting models is a little surprising. However, I think this may have an intuitive explanation.

We used BERT embeddings for all of our model predictions. BERT embeddings are very detailed, context aware vector representations of text data. This means that the features in the vector space were probably already pretty well sorted by target class value, and exhibit a high degree of linear separability. That is, thanks to the detailed nature of BERT embeddings there probably exists a straight line that can cut right down through the features vector space and do a pretty good job of keeping one target class on one side and the other target class on the other. In these conditions a cruder tool like logistic regression may have some advantages over tools which have the capacity to pickup complex relationships between features like gradient boosting algorithms.

Because the feature vectors were likely already well sorted, the adjustments the logistic regression algorithm must make to its weights are likely a function of where those feature vectors stand in relation to the best line of demarcation between them and the feature vectors of the other target value. Therefore, instead of having to learn complex relationships between the features, the model can just say, "when the feature vector is x distance from most of the feature vectors of the other target value, I need to adjust my weights to compute a dot product that will result in a 1 (or a 0) when passed to the sigmoid function". This is a pattern logistic regression can easily pickup on during training.

Conversely, tree based algorithms like our gradient boosters carve up vector space in a very different way. These models delineate relationsips between features in space by making axis aligned adjustements. That is, if there is potential to decrease the loss function output when some condition is met, the model will either make a 90 degree turn or it will keep going stright. This means it is the carving up of space in right angles which offers it more maneuverability. This is useful when there are complex relationships between features to be discerned. However, because BERT itself picks up on many of these relationships and places the feature vectors accordingly, this maneuverability of the gradient boosting models is less useful. Moreover, the closer the two target value classes are to being linearly separable, the bigger a problem axis aligned splits become. This is because the more linearly separable the classes are the better a straight line does at approxiating the split, and a model that carves space up into right angles will struggle to approxiate a straight line. It can approximate a straight line if it makes a lot of really tiny splits (smaller right angle lines will appear straighter the smaller they get), but this allows the model extra maneuverability to fit every nook and cranny of feature relationships it can find in the training set. That is, its propensity to fit noise increases as it decreases the size of its splits in the hopes of approximating that straight line. Therefore, gradient boosting algorithms are stuck between grossly misrepresenting relationships between features because its right angles struggle to represent linear separability, or it will split so many times in the hope of approximating a linearly separable relationship that it will gain the maneuverability to fit relationships between features that do not generalize beyond the training set (i.e. noise).

At the outset we should note that iterations for model_cat and n_estimators for model_light were set at 200, meaning that each model made 200 residual predictions which it added as a correction to the original prediction given by the log odds of a positive prediction. If the model were attempting to emulate a straight line as much as possible, we would expect the highest value of iterations for model_cat and n_estimators for model_light to be chosen during hyperparameter tuning, and we would expect the highest value of depth for model_cat and max_depth for model_light to be chosen, as both iterations/n_estimators and tree depth per estimator allow the model to approximate a straight diagnol line more closely the more there are of them. However, since iterations and n_esimators are both statically set to 200, we would expect the selected values for depth, and any other hyperparameters that influence the number of splits made, to allow for the most splits possible given the hyperparameter values offered to the models. 

It is also worth noting that learning rate is orthognal to the model's ability to approximate a straight diagnol line in this case. While a smaller learning rate would require more iterations/n_estimators to reach a minimum loss value, the number of iterations/n_estimators in this case is set statically. Therefore, a smaller learning rate would not lead to more iterations/n_estimators, and to the extent that hyperparameters do impact the model's ability to fit a straight line, it will be by increasing or decreasing the number of splits (depth and max_depth values) for each estimator. All of the other hyperparameters are relevant in this regard.

For model_cat all of the hyperparameter values except for depth are chosen in a way that would allow for the maximum number of splits. If we interpret this in terms of the gradient boosting paradigm we layed out in the previous markdown cell, we would say that a depth of 8 allowed for too much overfitting during cross validation, such that the best scores were acheived with a depth of 3 even though this means the model would have been less able to approximate a straight diagnol line. We would also note that since all of the other hyperparameter values selected have the effect of maximizing splitting, the model likely reached the maximmim depth value frequently, if not on every iteration. If this is the case, it suggests that the optimal depth value may have been between 3 and 8 since allowing depths to get to 8 presumably led to overfitting, and all of the other hyperparameters were set to maximize depths. If we do not find that the average depth value for model_cat was close to 3, then we can speculate 3 was an optimmal depth value because it would mean that the model frequently did better cutting off before 3 despite having hyperparameter values that should allow it to reach 3. It may, for example, mean that all nodes had converged on the same classification before getting to 3. We will check the average tree depth of model_cat in the next code cell.

model_light:
For model_light all of the hyperparameter values selected during tuning were hyperparameters which maximized the number of splits per tree (estimator). Therefore, we can speculate that the model did not hit an overfitting problem in its pursuit to construct a diagnol straight line for a decision boundary, the way we believe model_cat did. Like model_cat, we will check the average tree depth for model_light in su subsequent code cells. If the average depth of the trees in model_light is 12 or very close to 12, it would suggest that perhaps the model would benefit from a higher range of depth values to tune on. 

There is no method or attribute for accessing tree depth directly in CatBoostClassifier. However, we can save the trained model as a json file and then access the 'oblivous_trees' dictionary key. data is a dictionary where each key (including 'oblivious_trees') contains a list of dictionaries. One of the keys in this list of dictionaries is 'splits' so we can assign 'oblivious_tees' to the variable trees and then iterate over each tree in trees and find the length of the splits (depth) to calculate the average depth of all of the decision trees. 

```python
with open('models/model_cat.json', 'r') as f:
    data = json.load(f)

trees = data['oblivious_trees']
depths_cat = [len(tree['splits']) for tree in trees]
avg_depth = sum(depths_cat) / len(depths_cat)
print(f"Average tree depth: {avg_depth:.2f}")
```

This result suggests that the optimal depth parameter for catboost may be between 3 and 8. Therefore, on option worth considering for anybody looking to improve upon this project would be to set all other tuned hyperparameters to their current values, and tune on the depth value from values of 4 to 7. 

```python
model_light = joblib.load('models/model_light.pkl')


booster = model_light.booster_
trees = booster.dump_model()["tree_info"]

def get_tree_depth(tree_structure):
    if "left_child" not in tree_structure or "right_child" not in tree_structure:
        return 1
    return 1 + max(get_tree_depth(tree_structure["left_child"]),
                   get_tree_depth(tree_structure["right_child"]))

depths_light = [get_tree_depth(tree["tree_structure"]) for tree in trees]
avg_depth = sum(depths_light) / len(depths_light)

print(f"Average tree depth (LightGBM): {avg_depth:.2f}")
```

The average tree depth of 10.36 suggests that a depth of 12 was just about optimal for this model, given the other hyperparamters chosen during tuning, because it means that the model frequently utilized the full depth value but not always. Speculating again, it seems likely that the model utilized the full depth early on, but for later estimators the residual (gradient with respect to the previous prediction) had gotten small enough that those later estimators were not reaping any benefit from splitting all the way to max_depth. The only way to know for sure would be to plot the tree depths. We will do that in the following code cell. 

```python
# Bar chart: depth by tree index
plt.figure(figsize=(12, 4))
plt.bar(range(len(depths_light)), depths_light)
plt.xlabel("Tree Index")
plt.ylabel("Tree Depth")
plt.title("Tree Depths Across Boosting Iterations (LightGBM)")
plt.tight_layout()
plt.show()
```

Surprisingly, model_light actually reached max_depth more often in the second hundred iterations, continuing to the end. This suggests that residuals in later rounds may have been harder to correct, leading to more splits per attempt. It is hard to say whether or not the model was fitting signal or noise in these latter rounds. However, this output suggests that a max_depth of 12 is probably appropriate, and that tuning on a range of iterations greater than 200 may be worthwhile in the event the model is still fitting to signal in these latter rounds.
